# Handling outliers

```python
plt.figure(figsize=(10, 10))
for i,col in enumerate(numerical_stddata.columns):
    plt.subplot(3,3,i+1)
    plt.boxplot(std_data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
#using IQR
def removing_outliers(columns):
    for i in columns:
        lower_quantile=std_data[i].quantile(.25)
        upper_quantile=std_data[i].quantile(.75)

        IQR=upper_quantile-lower_quantile

        upper_boundary=upper_quantile+1.5*IQR
        lower_boundary=lower_quantile-1.5*IQR

        std_data[i]=np.where(std_data[i]>upper_boundary,upper_boundary,std_data[i])
        std_data[i]=np.where(std_data[i]<lower_boundary,lower_boundary,std_data[i])
removing_outliers(numerical_stddata)
```

```python
def boxplotting(numerical_stddata):
    plt.figure(figsize=(10, 10))
    for i, col in enumerate(numerical_stddata.columns):
        plt.subplot(3, 3, i+1)
        plt.boxplot(std_data[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
boxplotting(numerical_stddata)
```
