- interporation

```python
#applyimg interpolation
std_data_interpolated = std_data.interpolate(method='linear')
std_data_interpolated
```

- Filling

```python
#foward fill
std_data_forwardfill = std_data.fillna(method= 'ffill')
std_data_forwardfill
```

```python
std_data_backwardfill = std_data.fillna(method='bfill')
std_data_backwardfill
```

- Imputing

```python
#checking for skewness
numerical_stddata = std_data.select_dtypes(include=['number'])
numerical_stddata.skew()
```

```python
# for slightly skewed
numerical_stddata = std_data.select_dtypes(include=['number'])
std_data_mean = numerical_stddata.fillna(numerical_stddata.mean())
std_data_mean
```

```python
# negative skew
std_data_mean = numerical_stddata.fillna(numerical_stddata.median())
std_data_mean
```
